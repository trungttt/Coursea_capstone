```python
df_toronto=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)[0] #read the table directly from the website into dataframe using pandas
```

```python
df_toronto #The dataframe consists of three columns: PostalCode, Borough, and Neighborhood
```

```python
df_toronto_notassigned= df_toronto[df_toronto.Borough != 'Not assigned'] #Ignore cells with a borough that is Not assigned
df_toronto_notassigned
```

```python
#two rows will be combined into one row with the neighborhoods separated with a comma 
df_toronto_combined= df_toronto_notassigned.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(list).apply(lambda x:', '.join(x)).to_frame().reset_index()
df_toronto_combined
```

```python
#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough
for index, row in df_toronto_combined.iterrows():
    if row['Neighbourhood'] == 'Not assigned':
        row['Neighbourhood'] = row['Borough']
```
